````python
import os
import json
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr
import pandas as pd
from io import StringIO

from huggingface_hub import login
````

````python
load_dotenv(override=True)

openai_api_key = os.getenv('OPENAI_API_KEY')
if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set")

MODEL = "gpt-4o-mini"
openai = OpenAI()
````

````python
def generate_data_with_openai(topic, format):

    print(f"topic: {topic}, format: {format}")

    system_message = f"""You are a useful assistant helping generating test data for various projects.
    You always return data in the requested format, no other explanation is needed, just the pure data.
    A do not need the leading (eg. ```csv) and trailer (```) strings either!
    """
    user_prompt = f"Generate test data for {topic} trading in {format} formt"

    prompt =  [
    {"role": "system", "content": system_message},
    {"role": "user", "content": user_prompt}
     ]

    completion = openai.chat.completions.create(model=MODEL, messages = prompt)
    generated_data = completion.choices[0].message.content

    string_io_object = StringIO(generated_data)

    file_name= f".local_data/w3_exc_generated_openai.{format}"

    if format == 'json':
        df = pd.read_json(string_io_object)
        df.to_json(file_name)
    elif format == 'csv':
        df = pd.read_csv(string_io_object)
        df.to_csv(file_name)
    else:
        raise ValueError (f"Format has to be json or csv, {format} is not allowed")

    return("file was generated")
````

````python
import gradio as gr

def process_input(text_input, dropdown_selection):
    """
    This function processes the text input and dropdown selection.
    """
    return f"You entered: '{text_input}' and selected: '{dropdown_selection}'"

iface = gr.Interface(
    fn=generate_data_with_openai,
    inputs=[
        gr.Textbox(label="Generate data about"),
        gr.Dropdown(choices=["csv", "json"], label="Select a format")
    ],
    outputs="text",
    title="Gradio Demo with Text, Dropdown, and Button"
)

iface.launch()
````
